### Data Summary
- **6 Dimensions** — Producers, Shows, Networks, Content Task Types, Rights Holders, Platforms
- **6 Batch Facts** — Content Doc Events, Rights Clearance, Crew Wellness, Content Quality, Client Satisfaction, Production Performance
- **6 Event Facts** → Eventhouse — MAM Interactions, Metadata Entries, Approval Workflows, Content Handoffs, Delivery Alerts, Regulatory Reviews
- **5 Streams** → Eventhouse — MAM Activity, Rights Status, Delivery Tracking, QC Results, Audience Metrics

### Ontology
- **6 Entity Types:** Producer, Show, Network, ContentTaskType, RightsHolder, Platform
- **5 Relationship Types:** produces_for, airs_on, assigned_task, holds_rights, delivers_to
- **6 Contextualizations:** ContentDocEvent, MAMInteraction, ApprovalWorkflow, RightsClearance, DeliveryAlert, MetadataEntry

```python
# ════════════════════════════════════════════════════════════════════════
# SCHEMA INSPECTION HELPER
# ════════════════════════════════════════════════════════════════════════

def preview_table(table_name, base_path=CSV_BASE_PATH, rows=5):
    """Read a CSV and display schema + sample rows."""
    path = f"{base_path}/{table_name}.csv"
    df = spark.read.option("header", True).option("inferSchema", True).csv(path)
    print(f"\n{'\u2500'*60}")
    print(f"Table: {table_name}  |  {df.count()} rows  |  {len(df.columns)} columns")
    print(f"{'\u2500'*60}")
    df.printSchema()
    df.show(rows, truncate=False)
    return df

# Uncomment to preview specific tables:
# preview_table("dim_producers")
# preview_table("fact_content_doc_events")
# preview_table("stream_mam_activity")
```
