### Ontology
- **6 Entity Types:** AccountExecutive, Campaign, InventoryUnit, Market, Advertiser, Vendor
- **5 Relationship Types:** manages_campaign, advertises_in_market, uses_inventory, services_vendor, targets_audience
- **6 Contextualizations:** CampaignDocEvent, OMSInteraction, ChartingEvent, ProofApproval, WorkOrder, POPAlert

```python
# ════════════════════════════════════════════════════════════════════════
# SCHEMA INSPECTION HELPER
# ════════════════════════════════════════════════════════════════════════

def preview_table(table_name, base_path=CSV_BASE_PATH, rows=5):
    """Read a CSV and display schema + sample rows."""
    path = f"{base_path}/{table_name}.csv"
    df = spark.read.option("header", True).option("inferSchema", True).csv(path)
    print(f"\n{'─'*60}")
    print(f"Table: {table_name}  |  {df.count()} rows  |  {len(df.columns)} columns")
    print(f"{'─'*60}")
    df.printSchema()
    df.show(rows, truncate=False)
    return df

# Uncomment to preview specific tables:
# preview_table("dim_account_executives")
# preview_table("fact_campaign_orders")
# preview_table("stream_campaign_pacing")
```
