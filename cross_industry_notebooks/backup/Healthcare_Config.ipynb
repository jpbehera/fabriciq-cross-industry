### Data Summary
- **6 Dimensions** — Nurses, Patients, Hospital Units, Documentation Types, Medications, Diagnoses
- **6 Batch Facts** — Shifts, Patient Encounters, Care Plans, Burnout Surveys, Patient Satisfaction, Documentation Quality
- **6 Event Facts** → Eventhouse — Documentation Events, Medication Admin, Vital Signs, Assessments, EHR Interactions, Handoffs
- **5 Streams** → Eventhouse — RTLS Location, EHR Clickstream, Nurse Call Events, BCMA Scans, Clinical Alerts

### Ontology
- **6 Entity Types:** Nurse, Patient, HospitalUnit, DocumentationType, Medication, Diagnosis
- **5 Relationship Types:** assigned_to, admitted_to, has_diagnosis, cares_for, prescribes_for
- **6 Contextualizations:** DocumentationEvent, MedicationAdministration, VitalSignsRecording, AssessmentEvent, EHRInteraction, HandoffReport

```python
# ════════════════════════════════════════════════════════════════════════
# SCHEMA INSPECTION HELPER
# ════════════════════════════════════════════════════════════════════════

def preview_table(table_name, base_path=CSV_BASE_PATH, rows=5):
    """Read a CSV and display schema + sample rows."""
    path = f"{base_path}/{table_name}.csv"
    df = spark.read.option("header", True).option("inferSchema", True).csv(path)
    print(f"\n{'─'*60}")
    print(f"Table: {table_name}  |  {df.count()} rows  |  {len(df.columns)} columns")
    print(f"{'─'*60}")
    df.printSchema()
    df.show(rows, truncate=False)
    return df

# Uncomment to preview specific tables:
# preview_table("dim_nurses")
# preview_table("fact_documentation_events")
# preview_table("stream_ehr_clickstream")
```
